```python
#Part A (solution2):
TextFilePath1 = "C:\\Users\\JUN GUO\\Google Drive\\CS221\\Projects\\Project1\\PartB\\40-intersection-A"
TextFilePath2 = "C:\\Users\\JUN GUO\\Google Drive\\CS221\\Projects\\Project1\\PartB\\40-intersection-B"
def tokenize(TextFilePath): 
    #input: String<TextFilePath>, e.g., "C:\\Users\\JUN GUO\\Google Drive\\CS221\\Projects\\Project1\\Test-File-0.txt"
    #output: List<Token>
    print("Tokenize!")
    Token = []
    split = " ,.<>!?+-*()[]{}`~@#$"
    L=1
    #if(split[0]==" "):
    #    print(split)
    with open(TextFilePath) as f: 
        for line in f.readlines():
            word="";
            for i in range(len(line)):
                if line[i].isalpha() or line[i].isdigit():
                    word += line[i]
                else:
                    if len(word)>=L:
                        Token.append(word.lower())
                    word=""
    return Token
def computWordFrequencies(Token):
    #hash: hash table that is used to save tokens and their frequencies
    print("Computing Word Frequencies!")
    hash = {}
    for key in Token:
        if key in hash:
            hash[key]+=1
        else:
            hash[key]=1
    return hash
def printFrequencies(Frequencies):
    print("Printing Tokens with the frequencies in decreasing order!")
    Frequencies = sorted(Frequencies.items(), key=lambda x:x[1], reverse = True)
    for item in Frequencies:
        print(item[0] + ": " + str(item[1]))
Token = tokenize(TextFilePath1)
Frequencies=computWordFrequencies(Token)
printFrequencies(Frequencies)
```

```python
#Part B:
import time
def Intersection(file1,file2):
    Tokens1 = tokenize(file1)
    #print(Tokens1[0],Tokens1[1],Tokens1[2])
    Tokens2 = tokenize(file2)
    #print(Tokens2[0],Tokens2[1],Tokens2[2])
    Frequencies1=computWordFrequencies(Tokens1)
    count=0
    hash2 = {}
    for word in Tokens2:
        if (word in Frequencies1) and (word not in hash2):
            count+=1
            hash2[word]=1
            #print(word,"\n")
    print("The number of common words is %d" %count)
startTime = time.time()
Intersection(TextFilePath1,TextFilePath2)
print("Running Time: %s seconds" % (time.time() - startTime))
```

```python
#Part C
def wordFrequencyAdd(file1,file2):
    hash={}
    with open(file1) as f: 
        for line in f.readlines():
            pair=[]
            for word in line.split(", "):
                pair.append(word)
            if pair[0] in hash:
                hash[pair[0]]+=int(pair[1])
            else:
                hash[pair[0]]=int(pair[1])
    with open(file2) as f: 
        for line in f.readlines():
            pair=[]
            for word in line.split(", "):
                pair.append(word)
            if pair[0] in hash:
                hash[pair[0]]+=int(pair[1])
            else:
                hash[pair[0]]=int(pair[1])
    print("Writing the output file!")
    outFile = open("Output.txt", "w")
    for item in hash.items():
        outFile.write(item[0] + ", " + str(item[1]) + "\n")
    outFile.close()
FilePath1 = "C:\\Users\\JUN GUO\\Google Drive\\CS221\\Projects\\Project1\\PartC\\enwiki-1G.wc"
FilePath2 = "C:\\Users\\JUN GUO\\Google Drive\\CS221\\Projects\\Project1\\PartC\\enwiki-4000000.wc"
wordFrequencyAdd(FilePath1,FilePath2)
```
